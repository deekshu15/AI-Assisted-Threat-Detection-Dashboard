Task 1: CREATE DATABASE + CREATE TABLE Prompt

Act as an experienced SQL developer and create a SQL script that demonstrates the creation of a database and tables. Design a simple Student Management System database containing tables such as Students, Courses, and Enrollments.
```python
import sqlite3

# Create or connect to the SQLite database
conn = sqlite3.connect("StudentManagementSystem.db")
cursor = conn.cursor()

# Create Students table
cursor.execute("""
CREATE TABLE IF NOT EXISTS Students (
    StudentID INTEGER PRIMARY KEY AUTOINCREMENT,
    FirstName TEXT NOT NULL,
    LastName TEXT NOT NULL,
    Gender TEXT NOT NULL,
    DateOfBirth DATE NOT NULL,
    Email TEXT NOT NULL UNIQUE,
    Phone TEXT,
    AdmissionDate DATE DEFAULT CURRENT_DATE,
    Status TEXT NOT NULL DEFAULT 'Active'
);
""")

# Create Courses table
cursor.execute("""
CREATE TABLE IF NOT EXISTS Courses (
    CourseID INTEGER PRIMARY KEY AUTOINCREMENT,
    CourseName TEXT NOT NULL,
    CourseCode TEXT NOT NULL UNIQUE,
    Credits INTEGER NOT NULL DEFAULT 3,
    Department TEXT NOT NULL
);
""")

# Create Enrollments table
cursor.execute("""
CREATE TABLE IF NOT EXISTS Enrollments (
    EnrollmentID INTEGER PRIMARY KEY AUTOINCREMENT,
    StudentID INTEGER NOT NULL,
    CourseID INTEGER NOT NULL,
    EnrollmentDate DATE DEFAULT CURRENT_DATE,
    Grade TEXT,

    FOREIGN KEY (StudentID) REFERENCES Students(StudentID),
    FOREIGN KEY (CourseID) REFERENCES Courses(CourseID)
);
""")

conn.commit()

print("Student Management System database and tables created successfully.")
```

Task 6: HAVING Clause + JOINS Prompt

Act as an experienced SQL developer and create a SQL script demonstrating the HAVING clause and SQL JOIN operations. Use INNER JOIN and LEFT JOIN to retrieve related data from multiple tables. Apply GROUP BY with HAVING to filter grouped results. Include practical examples with comments explaining each query.

```python
# Query 1: INNER JOIN - Display student names with their enrolled courses
cursor.execute("""
SELECT
    Students.StudentID,
    Students.FirstName || ' ' || Students.LastName AS Student_Name,
    Courses.CourseName,
    Enrollments.Grade
FROM Students
INNER JOIN Enrollments
ON Students.StudentID = Enrollments.StudentID
INNER JOIN Courses
ON Enrollments.CourseID = Courses.CourseID;
""")

print("Student Enrollment Details (INNER JOIN):")
for row in cursor.fetchall():
    print(row)


# Query 2: LEFT JOIN - Display all students, including those not enrolled in any course
cursor.execute("""
SELECT
    Students.StudentID,
    Students.FirstName || ' ' || Students.LastName AS Student_Name,
    Courses.CourseName
FROM Students
LEFT JOIN Enrollments
ON Students.StudentID = Enrollments.StudentID
LEFT JOIN Courses
ON Enrollments.CourseID = Courses.CourseID;
""")

print("\nAll Students with Their Courses (LEFT JOIN):")
for row in cursor.fetchall():
    print(row)


# Query 3: Count the number of students enrolled in each course
cursor.execute("""
SELECT
    Courses.CourseName,
    COUNT(Enrollments.StudentID) AS Total_Students
FROM Courses
LEFT JOIN Enrollments
ON Courses.CourseID = Enrollments.CourseID
GROUP BY Courses.CourseID;
""")

print("\nNumber of Students Enrolled in Each Course:")
for row in cursor.fetchall():
    print(row)


# Query 4: Use HAVING to display only courses having more than one student
cursor.execute("""
SELECT
    Courses.CourseName,
    COUNT(Enrollments.StudentID) AS Total_Students
FROM Courses
LEFT JOIN Enrollments
ON Courses.CourseID = Enrollments.CourseID
GROUP BY Courses.CourseID
HAVING COUNT(Enrollments.StudentID) > 1;
""")

print("\nCourses with More Than One Student (HAVING):")
for row in cursor.fetchall():
    print(row)


# Query 5: Display students with the number of courses they have enrolled in
cursor.execute("""
SELECT
    Students.FirstName || ' ' || Students.LastName AS Student_Name,
    COUNT(Enrollments.CourseID) AS Total_Courses
FROM Students
LEFT JOIN Enrollments
ON Students.StudentID = Enrollments.StudentID
GROUP BY Students.StudentID;
""")

print("\nNumber of Courses Taken by Each Student:")
for row in cursor.fetchall():
    print(row)


# Query 6: Use HAVING to display only students enrolled in more than one course
cursor.execute("""
SELECT
    Students.FirstName || ' ' || Students.LastName AS Student_Name,
    COUNT(Enrollments.CourseID) AS Total_Courses
FROM Students
INNER JOIN Enrollments
ON Students.StudentID = Enrollments.StudentID
GROUP BY Students.StudentID
HAVING COUNT(Enrollments.CourseID) > 1;
""")

print("\nStudents Enrolled in More Than One Course:")
for row in cursor.fetchall():
    print(row)
```

```python
# ----------------------------------------------------------
# Demonstration of Database Constraints
# ----------------------------------------------------------

# Create a new table to demonstrate constraints
cursor.execute("""
CREATE TABLE IF NOT EXISTS Departments (
    DepartmentID INTEGER PRIMARY KEY AUTOINCREMENT,   -- PRIMARY KEY
    DepartmentName TEXT NOT NULL UNIQUE,              -- NOT NULL + UNIQUE
    Building TEXT NOT NULL,
    Floor INTEGER CHECK(Floor > 0)                    -- CHECK Constraint
);
""")

conn.commit()

print("Departments table created successfully!")


# ----------------------------------------------------------
# Insert sample data into Departments
# ----------------------------------------------------------

cursor.executescript("""
INSERT OR IGNORE INTO Departments (DepartmentName, Building, Floor)
VALUES
('Computer Science', 'Block A', 2),
('Information Technology', 'Block B', 3),
('Electronics', 'Block C', 1);
""")

conn.commit()

print("Department records inserted successfully!")


# ----------------------------------------------------------
# Subquery 1:
# Display students enrolled in CourseID = 1
# ----------------------------------------------------------

cursor.execute("""
SELECT FirstName, LastName
FROM Students
WHERE StudentID IN (
    SELECT StudentID
    FROM Enrollments
    WHERE CourseID = 1
);
""")

print("\nStudents Enrolled in Course ID 1:")
for row in cursor.fetchall():
    print(row)


# ----------------------------------------------------------
# Subquery 2:
# Display courses having at least one enrollment
# ----------------------------------------------------------

cursor.execute("""
SELECT CourseName
FROM Courses
WHERE CourseID IN (
    SELECT DISTINCT CourseID
    FROM Enrollments
);
""")

print("\nCourses with Enrolled Students:")
for row in cursor.fetchall():
    print(row)


# ----------------------------------------------------------
# Subquery 3:
# Display students enrolled in more than one course
# ----------------------------------------------------------

cursor.execute("""
SELECT FirstName, LastName
FROM Students
WHERE StudentID IN (
    SELECT StudentID
    FROM Enrollments
    GROUP BY StudentID
    HAVING COUNT(CourseID) > 1
);
""")

print("\nStudents Enrolled in More Than One Course:")
for row in cursor.fetchall():
    print(row)


# ----------------------------------------------------------
# Subquery 4:
# Display the course having the maximum credits
# ----------------------------------------------------------

cursor.execute("""
SELECT CourseName, Credits
FROM Courses
WHERE Credits = (
    SELECT MAX(Credits)
    FROM Courses
);
""")

print("\nCourse with Maximum Credits:")
for row in cursor.fetchall():
    print(row)


# ----------------------------------------------------------
# Display the structure of Departments table
# (Shows the applied constraints)
# ----------------------------------------------------------

cursor.execute("PRAGMA table_info(Departments);")

print("\nDepartments Table Structure:")
for row in cursor.fetchall():
    print(row)
```
